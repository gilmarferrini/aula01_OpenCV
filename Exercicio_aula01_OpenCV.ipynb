```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('image.jpg')

cv2_imshow(image)
```

```python
cv2.rectangle(image, (0, 0), (1200, 700), 0, 12)
cv2_imshow(image)
```

```python
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(imageGray)
```
